```python
import pandas as pd
```

```python
DF1 = pd.DataFrame({'Std_ID': [1, 2, 3, 4],
       'Marks': [10, 20, 30, 40],
       'Course': ['C', 'C++', 'Python', 'Java']})
DF1
```

```python
DF2 = pd.DataFrame({'Std_ID': [1,2,4,5,6],
       'Age': [22, 23, 18, 19, 20],
       'Class': ['H', 'E', 'M', 'S', 'S S']})

DF2
```

```python
# INNER JOIN
DF3 = DF1.merge(DF2)

DF3


# DF1 ---- Left
# DF2 ---- Right

```

```python
pd.merge(DF1, DF2)
```

```python
pd.merge(DF2, DF1)

# DF2 ---- Left
# DF1 ---- Right
```

```python
pd.merge(DF1, DF2, how = 'outer')
```

```python
pd.merge(DF2, DF1, how = 'outer')
```

```python
print(DF1)
print(DF2)
pd.merge(DF1, DF2, on = "Course")
```

Create 2 Datafarme
1. Std_Id, Course, Marks
2. Std_Id, Age, Course

merge using course

```python
DF1 = pd.DataFrame({'Std_Id': [1, 2, 3, 4],
       'Marks': [10, 20, 30, 40],
       'Course': ['C', 'C++', 'Python', 'Java']})
DF1
```

```python
DF2 = pd.DataFrame({'Std_Id': [1,2,30,40],
       'Age': [22, 23, 18, 19],
       'Course': ['C', 'C++', 'Maths', 'Hindi']})

DF2
```

```python
print(DF1)
print(DF2)
```

```python
pd.merge(DF1, DF2)
```

```python
pd.merge(DF1, DF2, on = 'Course')
```

```python
pd.merge(DF1, DF2, on = ['Course', 'Std_Id'])
```

```python
pd.merge(DF1, DF2, on = 'Course', suffixes = ['DF1', 'DF2'])
```

```python
pd.merge(DF1, DF2, on = 'Course', suffixes = ['DF1', 'DF2'], how = 'outer')
```

```python
pd.merge(DF1, DF2, on = 'Course', suffixes = ['DF1', 'DF2'], how = 'inner')
```

```python
pd.merge(DF1, DF2, on = 'Course', suffixes = ['DF1', 'DF2'], how = 'left')
```

```python
pd.merge(DF1, DF2, on = 'Course', suffixes = ['DF1', 'DF2'], how = 'right')
```
